# 2.) RevolvingUtilizationOfUnsecuredLines

```python
train['RevolvingUtilizationOfUnsecuredLines'].hist()
```

```python
sns.boxplot(x='SeriousDlqin2yrs',y='RevolvingUtilizationOfUnsecuredLines',data=train)
```

It seems as though we have some outliers. Let's remove the outliers.

```python
train['RevolvingUtilizationOfUnsecuredLines'].quantile([.975])
```

```python
train['RevolvingUtilizationOfUnsecuredLines'].quantile([.999])
```

The 99.9% quantile for this variable did not seem like an obvious outlier. Instead I decided to remove points above 25000 which was a visible outlier in the boxplot. 

```python
train=train[train['RevolvingUtilizationOfUnsecuredLines'] <25000]
```

# 3.) Age

```python
# Distribution of age

train['age'].hist()
```

```python
# It seems as if we have one point at 0 so we will remove this.

len(train[train['age']==0])
```

```python
train=train[train['age']>0]
```

```python
# Let's create a function to do a hist plot, a boxplot, and a kde plot

def three_plots(var):
    train[var].hist()
    plt.show()
    sns.boxplot(data=train,x='SeriousDlqin2yrs',y=var)
    sns.displot(data=train,x=var,hue='SeriousDlqin2yrs',kind='kde',fill=True)
```

```python
three_plots("age")
```

- Age follows a normal distribution.
- Borrowers that experienced a serious delinquency tend to be a bit younger in general.

# 4. DebtRatio

```python
sns.boxplot(data=train,x='SeriousDlqin2yrs',y='DebtRatio')
plt.gca().set_ylim(0, 2)
```

# 9. NumberOfTime30-59DaysPastDueNotWorse

```python
three_plots('NumberOfTime30-59DaysPastDueNotWorse')
```

There seems to be an outlier at 98 so I will remove that.

```python
train=train[train['NumberOfTime30-59DaysPastDueNotWorse']<90]
```

```python
train['NumberOfTime30-59DaysPastDueNotWorse'].hist()
```

```python
sns.displot(data=train,x='NumberOfTime30-59DaysPastDueNotWorse',hue='SeriousDlqin2yrs',kind='kde',fill=True)
plt.gca().set_xlim(0,5)
plt.gca().set_ylim(0,1)
```

```python
sns.displot(data=train[train['SeriousDlqin2yrs']==1],x='NumberOfTime30-59DaysPastDueNotWorse',kind='kde',fill=True)
plt.gca().set_xlim(0,5)
```

It's hard to tell with these graphs, if the number of times a person was late for a loan by 30-60 days has a higher chance of experiencing serious delinquency, so I will just calculate the mean for each.

```python
pd.DataFrame({'Not serious delinquency':train[train['SeriousDlqin2yrs']==0]['NumberOfTime30-59DaysPastDueNotWorse'].mean(),
             'Serious delinquency': train[train['SeriousDlqin2yrs']==1]['NumberOfTime30-59DaysPastDueNotWorse'].mean()},index=[0])
```

Okay there is some difference where people that have experienced being late for a loan by 30-60 days more times are more likely to experience serious delinquency in the future.
